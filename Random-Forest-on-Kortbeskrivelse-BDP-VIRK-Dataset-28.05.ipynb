```python
#Replace Kortbeskrivelse with APS (1) vs Non-APS (0)

df = df.replace({'IVS': 0})
df = df.replace({'K/S': 0})
df = df.replace({'APS': 1})
df = df.replace({'A/S': 0})
df = df.replace({'KAS': 0})
df = df.replace({'FAP': 0})
df = df.replace({'EFO': 0})
df = df.replace({'I/S': 0})
df = df.replace({'FAS': 0})
df = df.replace({'FMA': 0})
df = df.replace({'FBA': 0})
df = df.replace({'ABA': 0})
df = df.replace({'FIV': 0})
df = df.replace({'SCE': 0})
df = df.replace({'MSS': 0})
df = df.replace({'FOR': 0})
```

```python
#Overview of Status Counts

df['status'].value_counts()
```

```python
#Replace Status with 1 (Open or Closed, without issues) vs 0 (Closed, with issues)

df = df.replace({'Aktiv': 1})
df = df.replace({'NORMAL': 1})
df = df.replace({'OPLØST EFTER ERKLÆRING': 1})
df = df.replace({'OPLØST EFTER FRIVILLIG LIKVIDATION': 0})
df = df.replace({'OPLØST EFTER FUSION': 1})
df = df.replace({'OPLØST EFTER GRÆNSEOVERSKRIDENDE HJEMSTEDSFLYTNING': 1})
df = df.replace({'OPLØST EFTER KONKURS': 0})
df = df.replace({'OPLØST EFTER SPALTNING': 0})
df = df.replace({'SLETTET': 0})
df = df.replace({'TVANGSOPLØST': 0})
df = df.replace({'UDEN RETSVIRKNING': 0})
df = df.replace({'UNDER FRIVILLIG LIKVIDATION': 0})
df = df.replace({'UNDER KONKURS': 0})
df = df.replace({'UNDER REASSUMERING': 0})
df = df.replace({'UNDER REKONSTRUKTION': 1})
df = df.replace({'UNDER TVANGSOPLØSNING': 0})
```

```python
# Current Split of Non-Support-Status (1) vs Support-Status (0)

df['status'].value_counts()
```

```python
neg = len(df[df['status'] == 0])
pos = len(df[df['status'] == 1]) 
total = df['status'].count()
print(neg)
print(pos)
```

```python
weight_for_0 = (1 / neg)*(total)/2.0 
weight_for_1 = (1 / pos)*(total)/2.0

class_weight = {0: weight_for_0, 1: weight_for_1}

print('Weight for class 0: {:.2f}'.format(weight_for_0))
print('Weight for class 1: {:.2f}'.format(weight_for_1))
```

```python
#Covert Dateframe into Year, Month, and Day

df['date_time']=pd.to_datetime(df['stiftelsesDato'], format='%Y-%m-%d')

#Create two new columns, one for the year and one for the month
df['Day_year'] = df['date_time'].dt.year
df['Day_month'] = df['date_time'].dt.month
```

```python
#Specifying X and y Variables
X=df[['status','branchekode','kommuneKode','Day_year','Day_month']]
y=df['kortBeskrivelse'].values
```

```python
# Import Preprocessing
from sklearn import preprocessing

# Import TrainTestSplit
from sklearn.model_selection import train_test_split

#Do Split (80% Train, 20% Test)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn import tree

# Import BaggingClassifier
from sklearn.ensemble import BaggingClassifier

# Import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

# Instantiate DT
dt = DecisionTreeClassifier(random_state=1,class_weight = class_weight)
dt2 = tree.DecisionTreeRegressor(max_depth=3, min_samples_split=2, random_state=1)

# Instantiate BC
bc = BaggingClassifier(base_estimator=dt, n_estimators=1000, random_state=1)
bc
```

```python
from sklearn import tree
```

```python
dt2.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
classnames = ['status','Branchekode','kommuneKode','Day_year','Day_month']
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dt2,filled=True,feature_names=classnames)
plt.show()
```

```python
#Import AccuracyScore

from sklearn.metrics import accuracy_score

# Fit bc to Train Set
bc.fit(X_train, y_train)
# Predict Test Set
y_pred = bc.predict(X_test)
# Evaluate Accuracy
acc_test = accuracy_score(y_test, y_pred)
print('Test set accuracy of bc: {:.2f}'.format(acc_test))
```

```python
#Import ConfusionMatrix, Seaborn, Matplotlib and Numpy

from sklearn.metrics import confusion_matrix
import seaborn as sn
import numpy as np 
%matplotlib inline

#Plot ConfustionMatrix

data = confusion_matrix(y_test, y_pred)
df_cm = pd.DataFrame(data, columns=np.unique(y_test), index = np.unique(y_test))
df_cm.index.name = 'Actual'
df_cm.columns.name = 'Predicted'
plt.figure(figsize = (10,7))
sn.set(font_scale=1.4)#for label size
sn.heatmap(df_cm, cmap="Blues", annot=True,annot_kws={"size": 16})# font size
```

```python
# Import RandomForest

from sklearn.ensemble import RandomForestRegressor

# Instantiate RF
rf = RandomForestRegressor(n_estimators=100, random_state=1)
```

```python
# Fit RF to Train Set

rf.fit(X_train, y_train) 
```

```python
# Import MeanSquareError

from sklearn.metrics import mean_squared_error as MSE

# Predict Test Set
y_pred = rf.predict(X_test)

# Evaluate the test set MSE
rmse_test = MSE(y_test, y_pred)**(1/2)

# Print MSE
print('Test set MSE of rf: {:.2f}'.format(rmse_test))
```

```python
# Create a pd.Series of Feature Importance

importances = pd.Series(data=rf.feature_importances_,
                        index= X_train.columns)
# Sort Feature Importance

importances_sorted = importances.sort_values()

# Plot Feature Importance

importances_sorted.plot(kind='barh', color='blue')
plt.title('Feature Importance')
plt.show()
```
